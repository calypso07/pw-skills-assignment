##Q.3

App routing in Flask is a way of defining the URLs that will be associated with the application. App routes allow developers to map URLs to specific functions that will be executed when the URL is requested. This allows us to create dynamic websites by defining URLs that will trigger different functions and actions in order to generate content. By using app routes, we can also create URLs that are more user-friendly and easier to remember, as well as URLs that are more SEO friendly.

App routes are used to define the paths and endpoints within a web application. They are used to specify how users will interact with the application and what they will see when they navigate to certain URLs. App routes are used to create a logical structure for the application, connect resources to URLs, and provide a way to easily manage the application.

##Q.5

The url_for() function is used to build a URL to the specific function dynamically. The first argument is the name of the specified function, and then we can pass any number of keyword argument corresponding to the variable part of the URL.

This function is useful in the sense that we can avoid hard-coding the URLs into the templates by dynamically building them using this function.

Consider the following python flask script.

```python
from flask import Flask, redirect, url_for
app = Flask(__name__)

@app.route('/admin')
def hello_admin():
    return 'Hello Admin'

@app.route('/guest/<guest>')
def hello_guest(guest):
    return 'Hello %s as Guest' % guest

@app.route('/user/<name>')
def hello_user(name):
    if name =='admin':
        return redirect(url_for('hello_admin'))
    else:
        return redirect(url_for('hello_guest',guest = name))

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
